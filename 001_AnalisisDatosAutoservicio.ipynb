¡Define una get_item()función que reciba un parámetro, el número del artículo que deseas pedir, y devuelva el nombre de ese artículo!

Por ejemplo, si llamaste a la función con:

* Valor del argumento 1, podría devolver '🍔 Cheeseburger'.
* Valor del argumento 2, podría devolver '🍟 Fries'.
* Valor del argumento 3, podría devolver '🥤 Soda'.
* Valor del argumento 4, podría devolver '🍦 Ice Cream'.
* Valor del argumento 5, podría devolver '🍪 Cookie'.

¡Asegúrate de llamar a esta función varias veces para comprobar que funciona!

Por último, hagamos lo siguiente:

```python
def get_item(particulo):
  match particulo:
    case 1:
      lista_articulos.append("Chesseburger")
    case 2:
      lista_articulos.append("Fries")
    case 3:
      lista_articulos.append("Soda")
    case 4:
      lista_articulos.append("Ice cream")
    case 5:
      lista_articulos.append("Cookie")

lista_articulos=[]

while True:
  print("1 -🍔 Cheeseburger")
  print("2 -🍟 Fries")
  print("3 -🥤 Soda")
  print("4 - 🍦 Ice Cream")
  print("5 - 🍪 Cookie")
  articulo=int(input("Escribe el número del artículo (6 para salir): "))
  if articulo == 6:
    break
  else:
    get_item(articulo)
print("\nTu orden es:\n")
for item_comida in  lista_articulos:
  print(item_comida)

```
