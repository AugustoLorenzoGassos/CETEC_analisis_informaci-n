# Instrucciones
La gente dice que cada año de la vida de un perro equivale, aproximadamente, a siete años de la vida de un humano. 🥲

Escribe una función llamada dog_years() que reciba dos parámetros:

name (cadena de texto / string)

age (número entero / integer)

La función debe calcular y devolver una cadena de texto (frase) que incluya tanto el nombre del perro como su edad convertida a años humanos.

Ejemplo:
Si los perros Landon, Red y Cooper tienen 3, 6 y 2 años respectivamente, la función debería calcular sus edades como 21, 42 y 14 años humanos



```
dog_years('Landon', 3)
dog_years('Red Bean', 6)
dog_years('Cooper', 2)
```

Los mensajes devueltos (el return) deben verse así:

* Landon tiene 21 años en años humanos.

* Red Bean tiene 42 años en años humanos.

* Cooper tiene 14 años en años humanos.


````python
def dog_years(name,age):
  return f"{name} tiene {age} años en edad perro y {age*7} en edad humano"

print(dog_years("Landon",3))
print(dog_years("Red Bean",6))
print(dog_years("Cooper",2))





````
