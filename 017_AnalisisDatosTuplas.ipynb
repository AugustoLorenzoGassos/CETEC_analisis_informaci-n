#Tuplas
Las tuplas son colecciones ordenadas que no se pueden modificar después de su creación.

Esto significa que son inmutables , lo que implica que sus valores no pueden cambiar. Son ideales para datos que deben permanecer invariables, como coordenadas o valores de color.

Sin embargo, podemos reemplazar una tupla por otra y tener duplicados dentro de una misma tupla.

Las listas y las tuplas tienen algunas cosas en común:

* Pueden almacenar varios elementos en una sola variable.
* Sus valores están separados por comas.

Sin embargo, las tuplas se crean utilizando ( )paréntes

```python
#Tupla Valida
t1= ("hola",)
#Tupla Valida
t2= "Hola",
#Tupla invalida
t3= ("hola")
```

Gracias a internet, podemos conectar con amigos de todo el mundo. 🧑‍🤝‍🧑

Cuando tú y tus amigos estáis dispersos en diferentes ciudades, compartir ubicaciones puede ayudaros a sentiros más cerca. 🌐

¡Usemos la latitud y la longitud para crear tuplas con las ubicaciones de nuestros amigos!

* Crea una tupla para la ciudad en la que te encuentras.
* Crea 3 tuplas para tus amigos, cada una con la latitud y la longitud de sus ciudades.
* Imprime la ubicación de todos tus amigos.
¿Cuál de tus amigos está más lejos?

Bonus : Intenta combinar todas las tuplas en una sola.

```python
mi_ubicacion = (19.53124,-96.91589)
#Pachuca, hidalgo
ubicacion_amigo1 = (20.11697,-98.73329)
#CDMX
ubicacion_amigo2 = (19.42847,-99.12766)
#Guadalajara, jalisco
ubicacion_amigo3 = (20.67738,-103.34749)
#Calucular distancias entre amigos
#amigo 1
distancia_amigo1 = (((ubicacion_amigo1[0]-mi_ubicacion[0])**2)+((ubicacion_amigo1[1]-mi_ubicacion[1])**2))**0.5
print(distancia_amigo1)
#amigo 2
distancia_amigo2 = (((ubicacion_amigo2[0]-mi_ubicacion[0])**2)+((ubicacion_amigo2[1]-mi_ubicacion[1])**2))**0.5
print(distancia_amigo2)
#amigo 3
distancia_amigo3 = (((ubicacion_amigo3[0]-mi_ubicacion[0])**2)+((ubicacion_amigo3[1]-mi_ubicacion[1])**2))**0.5
print(distancia_amigo3)
print("\n")
distancias = [distancia_amigo1, distancia_amigo2, distancia_amigo3]

distancicia_mas_larga = distancias[0]
for item_distancia in distancias:
  if item_distancia > distancicia_mas_larga:
    distancicia_mas_larga = item_distancia

print(f"la distancia más alejada de mi punto es: {distancicia_mas_larga}")
todas_las_ubicaciones = (mi_ubicacion, ubicacion_amigo1, ubicacion_amigo2, ubicacion_amigo3)
print("\nUbiucaciones unidas")
print(todas_las_ubicaciones)
```

El Sistema de Logística Global
Imagina que trabajas en una empresa de envíos. Tienes una lista de "Entregas" donde cada entrega es una tupla con la siguiente estructura: (ID_Pedido, (Latitud, Longitud), "Nombre_Ciudad", Peso_KG).

Tu tarea es:

Crear una lista con al menos 4 tuplas de este tipo.

Filtrar y mostrar solo las ciudades que superen un peso de carga específico (por ejemplo, 15 kg).

Calcular el peso total de todas las entregas sumando el cuarto elemento de cada tupla.

Desempaquetado: Al iterar, desempaqueta la latitud y longitud para mostrarla por separado.

```python
peso_maximo = 3
peso_total = 0

lista_entregas = [("00001",(19.53124,-96.91589),"Xalapa, Veracruz",3),
("00002",(20.11697,-98.73329),"Pachuca, Hidalgo",7),
("00003",(19.42847,-99.12766),"CDMX",2),
("00003",(20.67738,-103.34749),"Guadalajara, Jalisco",9)]

lista_entrega_ubicaciones = []

for item_entregas in lista_entregas:
  if item_entregas[3]>peso_maximo:
    print(f"La entrega en la ciudad de {item_entregas[2]} con un peso de {item_entregas[3]} Kg. rebasa el peso máximo permitido {peso_maximo}")
  peso_total += item_entregas[3]
  lista_entrega_ubicaciones.append(item_entregas[1])

print(f"El peso total de todas las entregas es de {peso_total} kg")
print("Las ubicaciones de entrega son: ", lista_entrega_ubicaciones)

```
