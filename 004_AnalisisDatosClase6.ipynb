```python
adn = ["AUG","AUC","UCG"]

adn.append("UAA")
adn.insert(2,"GAU")

adn.remove("AUC")
adn.pop(0)

print(adn)
```

La diferencia entre las funciones integradas y los métodos en una lista radica en que los métodos utilizan la notación de punto en la variable de lista que creamos. Las funciones integradas se pueden llamar de forma independiente, pero los métodos siempre están asociados a la variable de lista desde la que se llaman.

Otra diferencia destacable es que, si bien no todas las funciones integradas están definidas para trabajar con listas, los métodos de lista solo funcionan con listas.
