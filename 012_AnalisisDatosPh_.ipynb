En química, el pH es una escala que se utiliza para especificar la acidez o basicidad de un líquido. 🧪

Crea un programa llamado ph_niveles.py que compruebe si un nivel de pH es básico, ácido o neutro.

Primero, crea un ph variable y pídele al usuario un valor entre 0 y 14.

Escribe una if declaración elif que else:

* Si ph es mayor que 7, imprime "Básico".
* Si ph es menor que 7, imprime "Ácido".
* De lo contrario, imprime "Neutral".

```python
ValorPh = float(input("Dame un valor entre 0 y 14: "))
if ValorPh >7:
  print("Basico")
elif ValorPh <7:
  print("Acido")

else:
  print("Neutral")
```
