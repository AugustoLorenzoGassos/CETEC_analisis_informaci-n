# Métodos de instancia
Ahora que podemos crear clases y objetos, y editar sus propiedades, nuestro último paso es crear funciones.

Espera... ¿funciones en clases? ¿No es eso súper complicado?

En realidad, no lo es. Las funciones que se definen en una clase se llaman métodos . De hecho, ya hemos utilizado varios métodos. Por ejemplo, en el ejercicio 25 , vimos algunos métodos de lista integrados, como:

* .insert()
* .append()
* .remove()

Estos son métodos integrados. ¡Vamos a crear nuestros propios métodos dentro de una clase!

Consideremos nuestra Studentclase:





```python
class Student:
  def __init__(self, name, year, enrolled, gpa):
    self.name = name
    self.year = year
    self.enrolled = enrolled
    self.gpa = gpa
```

Para implementar un método, necesitamos definirlo dentro de la clase. Por ejemplo, si queremos crear un método llamado .display_info(), lo definiríamos de la siguiente manera:

```python
class Student:
  def __init__(self, name, year, enrolled, gpa):
    self.name = name
    self.year = year
    self.enrolled = enrolled
    self.gpa = gpa
  def display_info(self):
    print("El estudiante " + self.name + " Su Gpa es "+ str(self.gpa))
```

```python
erick = Student("ERICK", 11, False, 4.0)
maria = Student("MARIA", 11, True, 3.8)

erick.display_info()
maria.display_info()
```

A continuación, vamos a crear tres métodos:

* .deposit(): Añadir dinero a la cuenta y devolver el nuevo balance.
* .withdraw(): Retirar dinero restando balancey devolviendo la cantidad retirada.
* .display_balance(): Imprime el valor actual de balance.

Por último, inicialice un nuevo objeto de la BankAccount clase y utilice estos métodos para realizar lo siguiente:

Deposita 96 dólares en la cuenta.
Retirar 25 dólares de la cuenta.
Imprime el saldo actual de la cuenta.

```python
class BankAccountclase:
  def __init__(self, first_name,last_name,account,account_type,pin,balance):
    self.fist_name = first_name
    self.last_name = last_name
    self.account = account
    self.account_type = account_type
    self.pin = pin
    self.balance=balance

  def deposit(self, cantidad_depositar):
    if cantidad_depositar<=0:
      print("Error: la cantidad a depositar debe ser mayor que 0")
    else:
      self.balance += cantidad_depositar

  def withdraw(self,cantidad_retirar):
    if cantidad_retirar > self.balance:
      print(f"La cantidad a retirar {cantidad_retirar} rebasa el saldo disponible")
    else:
      self.balance -= cantidad_retirar

  def display_balance(self):
    return f"El saldo de la cuenta {self.account} a nombre de {self.fist_name} {self.last_name} es: {self.balance}"

Cuenta1 = BankAccountclase("Augusto","Lorenzo",1234,"Nómina",0000,100)
print(Cuenta1.display_balance())

Cuenta1.deposit(95)
print(Cuenta1.display_balance())

Cuenta1.withdraw(25)
print(Cuenta1.display_balance())
```

Reto: El Gestor de Libros 📚
Crea un archivo llamado library_system.py.

1.Definir la clase Book Usa el método init() para establecer los siguientes atributos:

title (cadena)

author (cadena)

isbn (entero)

category (cadena)

is_available (booleano, debe empezar siempre en True)

price_per_day (flotante)

2.Crear tres métodos:

.borrow_book(): Si el libro está disponible (is_available es True), cambia el estado a False e imprime un mensaje de éxito. Si no, avisa que ya está prestado.

.return_book(): Cambia el estado a True e imprime que el libro ha sido devuelto.

.calculate_late_fee(days): Recibe un número de días de retraso y devuelve el costo total multiplicando days por price_per_day.

3.Ejecución: Inicializa un objeto de la clase Book con los datos que quieras y realiza lo siguiente:

Intenta pedir prestado el libro.

Calcula cuánto costaría una multa si alguien se retrasa 5 días en entregarlo.

Devuelve el libro.

Imprime el estado de disponibilidad fin

```python
from typing import Self
class Book:
  def __init__(self, title, author, isbn, category, is_available, price_per_day):
    self.title = title
    self.author = author
    self.isbn = isbn
    self.category = category
    self.is_available = True
    self.price_per_day = price_per_day

  def borrow_book(self):
    if self.is_available == True:
      self.is_available = False
      return f"El libro {self.title} fue prestado con éxito"
    else:
      return f"El libro {self.title} no esta disponible"

  def return_book(self):
    self.is_available = True
    return f"El libro {self.title} fue devuelto"

  def calculate_late_fee(self, days):
    return self.price_per_day*days

Libro = Book("Todo Muere","Juan Gómes Juagardo","1234","Trhiller",True,25)
print(Libro.borrow_book())
print(Libro.borrow_book())
print(Libro.return_book())
print(f"La cuenta por el atraso es de {Libro.calculate_late_fee(2)}")


```
