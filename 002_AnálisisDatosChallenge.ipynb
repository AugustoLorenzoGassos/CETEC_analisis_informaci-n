Un número se eleva al cuadrado cuando se multiplica por sí mismo o se eleva a la segunda potencia (por ejemplo, 4² = 4 x 4 = 16).

Primero, solicita al usuario un número entero con `int(input())` y guárdalo en una variable numérica. Luego, define una variable `total` con un valor inicial de 0.

Nota: Puedes pasar una cadena de texto a `int(input())`.

A continuación, usa un bucle `for` y la función `range()` para calcular la suma de los cuadrados de todos los números enteros desde 1 hasta ese número.

Finalmente, imprime el resultado como un valor entero.

Por ejemplo, si el número es 5, el total debería ser 55 porque:

$1 ^{2} +2^{2} + 3^{2} + 4^{2} + 5^{2} = 1 +4+9+16+25=55$

````python
NumeroFinal = 0
SumaTotal = 0

NumeroFinal = int(input("Dame un número fibal para calcular la serie: "))

for i in range(1,NumeroFinal+1,1):
  SumaTotal += i**2

print("Suma total: ", SumaTotal)
````
