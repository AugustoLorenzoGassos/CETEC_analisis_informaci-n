# Instrucciones
El Sombrero Seleccionador es un sombrero mágico parlante en la Escuela Hogwarts de Magia y Hechicería. El sombrero decide a cuál de las cuatro "Casas" pertenece cada estudiante de primer año:

* 🦁 Gryffindor
* 🦅 Ravenclaw
* 🦡 Hufflepuff
* 🐍 Slytherin

Escribe un programa que le haga algunas preguntas al usuario con las funciones int()y input()



```
P1) ¿Te gusta más el amanecer o el atardecer?

1) Amanecer

2) Atardecer
```
* Si la respuesta es igual a 1, tanto Gryffindor como Ravenclaw obtienen un +1.
* En caso contrario, si la respuesta es igual a 2, tanto Hufflepuff como Slytherin obtienen un +1.
* De lo contrario, muestra el mensaje "Entrada incorrecta".


```
P2) Cuando muera, quiero que me recuerden como:

1) El Bueno

2) El Grande

3) El Sabio

4) El Audaz
```
* Si la respuesta es 1, Hufflepuff +2.
* Si la respuesta es 2, Slytherin +2.
* De lo contrario, si la respuesta es 3, Ravenclaw +2.
* Si la respuesta es 4, Gryffindor +2.
* De lo contrario, muestra el mensaje "Entrada incorrecta".

```
P3) ¿Qué instrumento te resulta más agradable al oído?

1) El violín

2) La trompeta

3) El piano

4) La batería
```
* Si la respuesta es 1, Slytherin +4.
* Si la respuesta es 2, Hufflepuff +4.
* Si la respuesta es 3, Ravenclaw +4.
* Si la respuesta es 4, Gryffindor +4.
* De lo contrario, mostrará "Entrada incorrecta".

Por último, imprime la puntuación de cada casa.










````python
Gryffindor = 0
Ravenclaw = 0
Hufflepuff = 0
Slytherin = 0

Pregunta1 = int(input("¿Te gusta más el amanecer o el atardecer?: "))
Pregunta2 = int(input("Cuando muera, quiero que me recuerden como:"))
Pregunta3 = int(input("¿Qué instrumento te resulta más agradable al oído?: "))

if Pregunta1 == 1:
  Gryffindor += 1
  Ravenclaw += 1
elif Pregunta1 ==2:
  Hufflepuff +=1
  Slytherin +=1
else:
  print("Entrada incorrecta")

if Pregunta2 == 1:
  Hufflepuff +=2
elif Pregunta2 == 2:
  Slytherin +=2
elif Pregunta2 == 3:
  Ravenclaw += 2
elif Pregunta2 == 4:
  Gryffindor +=2
else:
  print("Entrada incorrecta")

if Pregunta3 == 1:
  Slytherin +=4
elif Pregunta3 == 2:
  Hufflepuff +=4
elif Pregunta3 == 3:
  Ravenclaw += 4
elif Pregunta3 == 4:
  Gryffindor +=4
else:
  print("Entrada incorrecta")

print("\nPuntajes por casa:\n")
print("Casa Gryffindor: ", Gryffindor)
print("Casa Ravenclaw: ", Ravenclaw)
print("Casa Hufflepuff: ", Hufflepuff)
print("Casa Slytherin: ", Slytherin)

````
