````python
PesosColombianos = int(input("Cantidad de pesos colombianos:"))
Soles = int(input("Cantidad de soloes peruanos:"))
Reales = int(input("Cantidad de reales brasileños:"))

Dolares = (PesosColombianos*0.000270165) + (Soles* 0.291) + (Reales * 0.19 )

print(f"La cantidad de dolaes es: {Dolares}")
````
