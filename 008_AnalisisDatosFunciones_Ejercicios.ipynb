# Instrucciones
¿Por qué hay tantos emojis de la Luna? Cada uno representa una fase lunar diferente, que es la forma de la parte de la Luna iluminada por el Sol vista desde la Tierra.

Escribe una función llamada moon_phase() que reciba un parámetro phase (el nombre de la fase lunar) y devuelva el emoji correcto según la siguiente lista:

* Luna Nueva	🌑
* Luna Creciente	🌒
* Cuarto Creciente	🌓
* Gibosa Creciente	🌔
* Luna Llena	🌕
* Gibosa Menguante	🌖
* Cuarto Menguante	🌗
* Luna Menguante	🌘


Nota: Si se pasa un nombre de fase inválido a la función moon_phase(), esta debe devolver el mensaje: 'Invalid moon phase' (Fase lunar inválida).

Llama a la función moon_phase() para probar que funcione correctamente.

```python
def moon_phase(pfase_lunar):
  match pfase_lunar.lower():
    case "luna nueva":
      print("Luna Nueva	🌑\n")
    case "luna creciente":
      print("Luna creciente 🌒\n")
    case "cuarto creciente":
      print("Cuarto creciente 🌓\n")
    case "gibosa creciente":
      print("Gibosa creciente 🌔\n")
    case "luna llena":
      print("Luna llena 🌕\n")
    case "gibosa menguante":
      print("Gibosa menguante 🌖\n")
    case "cuarto menguante":
      print("Cuarto menguante 🌗\n")
    case "luna menguante":
      print("Luna menguante 🌘\n")

while True:
  fase_lunar = input("Escribe la pase lunar para consultar la imagen ('Salir pata abandonar la consulta')")
  if fase_lunar.lower()=="salir":
    break
  else:
    moon_phase(fase_lunar)
```
