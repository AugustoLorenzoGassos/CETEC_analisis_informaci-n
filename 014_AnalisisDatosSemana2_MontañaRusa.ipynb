# Instrucciones
Desde 1927, la montaña rusa "The Cyclone" ha deleitado a los visitantes de Coney Island (Brooklyn, NY). 🎢

Están instalando un nuevo sistema de acceso (la altura mínima requerida es de 137 cm y el coste es de 10 créditos) ¡y necesitan tu ayuda para escribir el programa!

Crea un nuevo archivo llamado the_cyclone.

Pregúntale al usuario cuál es su altura y cuántos créditos tiene, y almacena los valores en una heightvariable y otra creditsvariable.

Utilice una combinación de operadores relacionales y lógicos para crear las reglas:

* Si son lo suficientemente altos y tienen suficientes créditos, imprime "¡Disfruta del viaje!"
* De lo contrario, si tienen suficientes créditos, pero no son lo suficientemente altos, imprime "No eres lo suficientemente alto para subirte".
* De lo contrario, si son lo suficientemente altos, pero no tienen suficientes créditos, imprime "No tienes suficientes créditos".
* De lo contrario, imprima un mensaje que indique que no han cumplido con ninguno de los dos requisitos.

```python
heightvariable = int(input("Escribe tu estatura (en centímetros): "))
creditsvariable = int(input("Escribe la cantidad de tus créditos: "))

if heightvariable<137 and creditsvariable<10:
  print("\nNo tienes la estatutra ni los créditos suficientes")
else:
  if heightvariable>=137 and creditsvariable>=10:
    print("\n¡Disfruta el viaje!")
  else:
    if creditsvariable>=10:
      print("\nNo eres lo suficientemente alto para subirte")
    else:
      print("\nNo tienes los suficientes creditos")
```
