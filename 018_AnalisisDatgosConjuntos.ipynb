El Reto: El Inventario del Smoothie Bar 🥤 Tienes tres proveedores distintos para tu negocio de smoothies. Necesitas saber exactamente qué frutas son exclusivas y cuáles están causando redundancia.

Tus conjuntos son:

Proveedor A: Manzana, Pera, Plátano, Arándano.

Proveedor B: Arándano, Tomate, Kiwi, Fresa.

Proveedor C: Fresa, Mango, Manzana, Guayaba.

Tu tarea en Python:

Encontrar la Unión de los tres proveedores (el inventario total).

Encontrar las frutas que solo tiene el Proveedor A y nadie más.

Encontrar la Diferencia Simétrica entre el Proveedor A y el Proveedor B (frutas que están en uno o en otro, pero no en ambos).


```python
proveedor_a = {"Manzana", "Pera", "Plátano", "Arándano"}
proveedor_b = {"Arándano", "Tomate", "Kiwi", "Fresa"}
proveedor_c = {"Fresa", "Mango", "Manzana", "Guayaba"}

union_proveedores = proveedor_a.union(proveedor_b, proveedor_c)
print("Unión de proveedores: ",union_proveedores)

frutas_exclusivas_a = proveedor_a.difference(proveedor_b, proveedor_c)
print("Furas exclusivas del proveedor a: ",frutas_exclusivas_a)
```
